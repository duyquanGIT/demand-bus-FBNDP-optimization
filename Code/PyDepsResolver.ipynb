```python
# @title 2. Project Configuration { display-mode: "form" }

# @markdown **Python Version:** (e.g., 3.12, 3.11, or leave empty for auto)
PYTHON_VERSION = "3.12.13" # @param {type:"string"}

# @markdown **Library List:** (Separate libraries with a comma, e.g.: numpy==1.26.4, pandas>=2.2, torch, xgboost)
LIBRARIES_INPUT = "numpy>=2, pandas, scikit-learn, xgboost, imbalanced-learn, shap, matplotlib, torch, pytorch-tabnet, optuna, joblib" # @param {type:"string"}

# Process the input: split by comma, trim whitespace, and ignore empty strings
libraries = [lib.strip() for lib in LIBRARIES_INPUT.split(',') if lib.strip()]

# Create the requirements.in file
with open("requirements.in", "w") as f:
    for lib in libraries:
        f.write(lib + "\n")

print(f"Success: 'requirements.in' created with {len(libraries)} libraries.")
print("--- Content of requirements.in ---")
!cat requirements.in
```

```python
# @title 3. Resolve and Export { display-mode: "form" }
# @markdown This will generate the final 'requirements.txt' and trigger a download.

import os
from google.colab import files

# Build the uv compile command
cmd = "uv pip compile requirements.in -o requirements.txt"
if PYTHON_VERSION.strip():
    cmd += f" --python-version {PYTHON_VERSION.strip()}"

print(f"Executing: {cmd}\n")

# Run the command
!{cmd}

print("\n" + "="*50)
if os.path.exists("requirements.txt"):
    print("FINAL RESOLVED REQUIREMENTS.TXT:\n")
    with open("requirements.txt", "r") as f:
        print(f.read())

    print("="*50)
    print("\nDownloading files...")
    files.download('requirements.in')
    files.download('requirements.txt')
else:
    print("ERROR: requirements.txt was not generated. Please check your library names.")
```
